## Perform basic data analysis

### Subtask:
Calculate basic statistics and perform groupings to identify patterns.


**Reasoning**:
Calculate the requested statistics, perform the grouping, and calculate the correlation matrix.



```python
# Calculate mean, median, and standard deviation for 'median_house_value'
median_house_value_mean = df['median_house_value'].mean()
median_house_value_median = df['median_house_value'].median()
median_house_value_std = df['median_house_value'].std()

print(f"Mean of median_house_value: {median_house_value_mean}")
print(f"Median of median_house_value: {median_house_value_median}")
print(f"Standard deviation of median_house_value: {median_house_value_std}")

# Group by 'housing_median_age' and calculate the average 'median_house_value'
avg_median_house_value_by_age = df.groupby('housing_median_age')['median_house_value'].mean()
print("\nAverage median_house_value by housing_median_age:")
display(avg_median_house_value_by_age)

# Find the minimum and maximum 'median_income'
min_median_income = df['median_income'].min()
max_median_income = df['median_income'].max()

print(f"\nMinimum median_income: {min_median_income}")
print(f"Maximum median_income: {max_median_income}")

# Calculate the correlation matrix
correlation_matrix = df.corr()
print("\nCorrelation matrix:")
display(correlation_matrix)
```

## Visualize the data

### Subtask:
Create different types of visualizations to represent the data and your findings.


**Reasoning**:
Create the required visualizations to represent the data and findings, including a line plot, bar plot, histogram, and scatter plot, with appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Line plot of average median_house_value by housing_median_age
plt.figure(figsize=(12, 6))
plt.plot(avg_median_house_value_by_age.index, avg_median_house_value_by_age.values)
plt.title('Average Median House Value by Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Average Median House Value')
plt.grid(True)
plt.show()

# 2. Bar plot showing total households for selected housing median ages
# Select a few interesting ages based on the line plot or general interest
selected_ages = [1, 15, 30, 52] # Example ages showing variations
selected_data = df[df['housing_median_age'].isin(selected_ages)]
households_by_age = selected_data.groupby('housing_median_age')['households'].sum()

plt.figure(figsize=(8, 6))
plt.bar(households_by_age.index.astype(str), households_by_age.values, color='skyblue')
plt.title('Total Households for Selected Housing Median Ages')
plt.xlabel('Housing Median Age')
plt.ylabel('Total Households')
plt.show()

# 3. Histogram of median_income
plt.figure(figsize=(8, 6))
plt.hist(df['median_income'], bins=50, color='lightcoral', edgecolor='black')
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()

# 4. Scatter plot of median_income vs. median_house_value
plt.figure(figsize=(10, 6))
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, color='green')
plt.title('Median House Value vs. Median Income')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()
```
